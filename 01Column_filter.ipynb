이 python 파일은 아무 데이터도 없는 Colunm을 제거하는 파일입니다.

```python
import numpy as np
import pandas as pd
```

```python
# Excel 파일 맨 밑의 행에 count함수를 사용하여 데이터가 없는 column은 0이 나오도록 만든 후 저장
#df = pd.read_excel(r'C:\Users\User\Desktop\projectss\01for_column_filter.xlsx',sheet_name = 'Sheet1')
df = pd.read_excel('01for_column_filter.xlsx',sheet_name = 'Sheet1')
```

```python
# Excel 데이터 rows 개수 세기
df.shape[0]
len(df.index)
```

```python
# Excel 데이터 columns 개수 세기
df.shape[1]
len(df.columns)
```

```python
for i in range(0,len(df.columns)):
    x = df.loc[len(df.index)-1,"Unnamed: %d" %i]
    if x == 0:
        df=df.drop("Unnamed: %d" %i, axis = 1)
```
