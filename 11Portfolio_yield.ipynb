```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_excel('11for_portfolio_yield.xlsx',sheet_name = 'Sheet1')
df2 = pd.read_excel('11for_portfolio_yield2.xlsx',sheet_name = 'Sheet1')
```

```python
df1 = df.fillna(0)

P1 = []
P2 = []
P3 = []
P4 = []
P5 = []

for i in range(1,len(df.columns)):
    if df1.loc[0, 'Unnamed: %d' %i] == 0:
        next
    else:
        P1.append(df1.loc[0, 'Unnamed: %d' %i])
        
for i in range(1,len(df.columns)):
    if df1.loc[1, 'Unnamed: %d' %i] == 0:
        next
    else:
        P2.append(df1.loc[1, 'Unnamed: %d' %i])
        
for i in range(1,len(df.columns)):
    if df1.loc[2, 'Unnamed: %d' %i] == 0:
        next
    else:
        P3.append(df1.loc[2, 'Unnamed: %d' %i])
        
for i in range(1,len(df.columns)):
    if df1.loc[3, 'Unnamed: %d' %i] == 0:
        next
    else:
        P4.append(df1.loc[3, 'Unnamed: %d' %i])
        
for i in range(1,len(df.columns)):
    if df1.loc[4, 'Unnamed: %d' %i] == 0:
        next
    else:
        P5.append(df1.loc[4, 'Unnamed: %d' %i])
```

```python
t = int(input('시점 t를 입력하세요(앞 코딩에서의 t를 입력할 것) : '))
K = int(input('보유 개월수 K를 입력하세요 : '))
df_new = pd.DataFrame()
for stock in P1:
    df_new[stock] = df2[stock]
print(len(df_new.columns))
for stock in P2:
    df_new[stock] = df2[stock]
print(len(df_new.columns))
for stock in P3:
    df_new[stock] = df2[stock]
print(len(df_new.columns))
for stock in P4:
    df_new[stock] = df2[stock]
print(len(df_new.columns))
for stock in P5:
    df_new[stock] = df2[stock]
print(len(df_new.columns))

df_new2 = df_new.pct_change(periods = t+K)

```
