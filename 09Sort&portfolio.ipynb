Portfolio를 구성하기 위해 list를 설정 후 sort로 value의 크기에 따라 내림차순하여 정렬하였습니다.

나머지 데이터가 1개인 경우 P1에 분배하였고 2개인 경우 P1과 P5에, 3개 이상인 경우 P1에 2개를 분배하고 나머지를 P5에 분배하도록 하였습니다.

for_portfolio.xlsx

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('09for_portfolio.xlsx',sheet_name = 'Sheet1')
df2 = pd.DataFrame()
```

```python
a = [0]*len(df.columns)
for i in range(0,len(df.columns)):
    a[i] = df.loc[0, "Unnamed: %d" %i], df.loc[1, "Unnamed: %d" %i]

def getKey(tup) :
    return tup[1]

a.sort(key=getKey, reverse=False)
```

```python
print(a)
```

```python
if len(df.columns)%5 == 0 :
    P1 = a[0:int(len(df.columns)/5)]
    P2 = a[int(len(df.columns)/5):int(len(df.columns)/5)*2]
    P3 = a[int(len(df.columns)/5)*2:int(len(df.columns)/5)*3]
    P4 = a[int(len(df.columns)/5)*3:int(len(df.columns)/5)*4]
    P5 = a[int(len(df.columns)/5)*4:int(len(df.columns))]
elif len(df.columns)%5 <= 2 :
    P1 = a[0:int(len(df.columns)/5)+1]
    P2 = a[int(len(df.columns)/5)+1:int(len(df.columns)/5)*2+1]
    P3 = a[int(len(df.columns)/5)*2+1:int(len(df.columns)/5)*3+1]
    P4 = a[int(len(df.columns)/5)*3+1:int(len(df.columns)/5)*4+1]
    P5 = a[int(len(df.columns)/5)*4+1:int(len(df.columns))]
else:
    P1 = a[0:int(len(df.columns)/5)+2]
    P2 = a[int(len(df.columns)/5)+2:int(len(df.columns)/5)*2+2]
    P3 = a[int(len(df.columns)/5)*2+2:int(len(df.columns)/5)*3+2]
    P4 = a[int(len(df.columns)/5)*3+2:int(len(df.columns)/5)*4+2]
    P5 = a[int(len(df.columns)/5)*4+2:int(len(df.columns))]
```

```python
df2.loc[1, "Unnamed: %d" %0] = 'P1'
df2.loc[2, "Unnamed: %d" %0] = 'P2'
df2.loc[3, "Unnamed: %d" %0] = 'P3'
df2.loc[4, "Unnamed: %d" %0] = 'P4'
df2.loc[5, "Unnamed: %d" %0] = 'P5'

for j in range(0,len(P1)):
    df2.loc[1, "Unnamed: %d" %(j+1)] = P1[j][0]
for j in range(0,len(P2)):
    df2.loc[2, "Unnamed: %d" %(j+1)] = P2[j][0]
for j in range(0,len(P3)):
    df2.loc[3, "Unnamed: %d" %(j+1)] = P3[j][0]
for j in range(0,len(P4)):
    df2.loc[4, "Unnamed: %d" %(j+1)] = P4[j][0]
for j in range(0,len(P5)):
    df2.loc[5, "Unnamed: %d" %(j+1)] = P5[j][0]
```

```python
df2.to_excel('09Real_portfolio.xlsx')
```
